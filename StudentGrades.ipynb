```python
#neo column total score
df['Total_Score'] = (df['MathScore'] + df['ReadingScore'] + df['WritingScore'])/3
```

```python
df[['Total_Score']]
```

```python
df['Total_Score'] = df['Total_Score'].round(2)
```

```python
sns.histplot(data=df,x='Total_Score', bins=70)
plt.title('Total Score Distribution')
plt.ylabel('Number of Scores')
plt.show()
```

```python
#We will create a new column Grade to categorize the students Scores
bins = [0, 60, 70, 80, 90, 100]
labels = ['Fail', 'Average', 'Good', 'Very Good', 'Excellent']
df['Grade'] = pd.cut(df['Total_Score'], bins=bins, labels=labels, right=False, ordered =True)
```

```python
df['Grade'].isna().sum()
```

```python
df['Grade'] = df['Grade'].fillna('Excellent')
```

**Conclusions**

Students who studied 5–10 hours weekly achieved the best grades. Both understudying and overstudying led to worse performance.

Students with higher-educated parents had higher average scores.

The students that practiced sports had better grades in average that those who didn't practice sports often or at all.

Female students achieved higher average grades compared to male students.
